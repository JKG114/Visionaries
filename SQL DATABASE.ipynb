Fact_Table (trip table):[{bikeid,starttime} (Primary Key) ,endtime,start_station_id (references station table), end station id (references station table) ,  {Date, Time} (references weather table), usertype, birth year, gender, trip duration]

Station_Table : [station id (primary key) , station name , station latitude, station longitude]

Weather_Table: [{Date, Time} (Primary Key), DewPointC, DewPointF, FeelsLikeC, FeelsLikeF, WindChillC, WindChillF, WindGustKmph, WindGustMiles,Cloud_Cover, humidity, PrecipMM, Pressure, TempC, TempF, Visibility, WeatherDesc, Wind_Dir_16_Point, Wind_Dir_Degree, WindSpeedKmph, WindSpeedMiles, ]

```python
def create_database(conn):
    """
    Using the database connection conn, create tables appropriate for 
    storing and using IMDB data.  Drops any pre-exsiting tables that would be re-defined by this procedure.
    """
    c = conn.cursor()
    c.execute('''DROP TABLE IF EXISTS weather''')
    c.execute('''DROP TABLE IF EXISTS station''')
    c.execute('''DROP TABLE IF EXISTS fact_table''')
    
    # Create database with tables.

    c.execute('''
           CREATE TABLE station(
                station_id int,
                station_name text,
                station_latitude float,
                station_longitute float,
                PRIMARY KEY(station_id)) ''') 
    
    c.execute('''
            CREATE TABLE weather(
                Date,
                Time,
                DewPointF, 
                FeelsLikeF,
                WindChillF,
                WindGustMiles,
                Cloud_Cover,
                humidity,
                PrecipMM, 
                Pressure,
                TempF,
                Visibility,
                WeatherDesc,
                Wind_Dir_16_Point,
                Wind_Dir_Degree,
                WindSpeedMiles, 
                PRIMARY KEY(Date, Time) )''')
    
#Took out Date, Time, FOREIGN KEY(Date, Time) REFERENCES weather(Date, Time),

    c.execute('''
            CREATE TABLE fact_table(
                bike_id int,
                starttime,
                endtime,
                start_station_id int,
                end_station_id int,    
                usertype text, 
                birth_year int, 
                gender int,
                trip_duration,
                Date,
                Time,
                PRIMARY KEY(bike_id, starttime, Date),
                FOREIGN KEY(Date, Time) REFERENCES weather(Date, Time),
                FOREIGN KEY(start_station_id) REFERENCES station(station_id), 
                FOREIGN KEY(end_station_id) REFERENCES station(station_id) )''') 
    # Commit all changes
    conn.commit()
    
# Call the create_database() function
create_database(conn)
print('done')
```

```python
import pandas

def populate_database(conn, csv_file):    
    reader = csv.reader(open(csv_file, encoding='latin1'))
    
    # skip header    
    next(reader, None)
    # Convert row to lower case
    for row in reader:
        row = [item.lower() for item in row]
        
        # insert into dimension tables
        # popuate directors dimension table with corresponding rows 
        try:
            c.execute('''
                INSERT INTO station (station_id, 
                station_name,
                station_latitude,
                station_longitude) VALUES (?,?,?,?)
                ''', (row[3], row[4],row[5],row[6]))
  
        except sqlite3.Error as e:
            conn.commit()
        
        # popuate actors dimension table with rows corresponding to actor1 
        try:
            c.execute('''
                INSERT INTO station (station_id int, 
                station_name text,
                station_latitude float,
                station_longitude float) VALUES (?,?,?,?)
                ''', (row[7], row[8],row[9],row[10]))
        
        except sqlite3.Error as e:
            conn.commit()
        
        try:
            c.execute(''' 
                INSERT INTO weather (Date,
                Time,
                DewPointF, 
                FeelsLikeF,
                WindChillF,
                WindGustMiles,
                Cloud_Cover,
                humidity,
                PrecipMM, 
                Pressure,
                TempF,
                Visibility,
                WeatherDesc,
                Wind_Dir_16_Point,
                Wind_Dir_Degree,
                WindSpeedMiles) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)
            ''', (row[21],row[22],row[24],row[26],row[28],row[30],row[31],row[32],row[33],row[34],row[36],row[37],
                  row[38],row[39],row[40],row[42]))
        except sqlite3.Error as e:
            conn.commit()

        try:
       
            c.execute('''
                INSERT INTO fact_table (bike_id, starttime, endtime, start_station_id, end_station_id, usertype, 
                birth_year, gender, trip_duration, Date, Time) VALUES (?,?,?,?,?,?,?,?,?,?,?)''', 
                      (row[11], row[15], row[16], row[3], row[7], row[12],row[13],row[14],row[2],row[21],row[22]))
        except sqlite3.Error as e:
#             print("fact_table, insert error:", e.args[0])

#     Commit to the connection in order to save the changes
            conn.commit()    
            
populate_database(conn,'Merged Data/Merged_January_2016_Bike.csv')
```
